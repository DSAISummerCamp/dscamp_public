# Named entity extraction

[List of all named entities](https://texthero.org/docs/api/texthero.nlp.named_entities)

```python
text = "November 3rd 2019. In the Kingdom of Python, Hani Safadi is king"
s = pd.Series(text)
s
```

```python
texthero.named_entities(s)[0]
```

```python
article = """ (CNN) - California's Death Valley is known to be a hot place, but it hit 130 degrees Fahrenheit (54.4 degrees Celsius) Friday for only the fifth time in recorded history -- that's only five days out of more than 40,000 days on record.
Interestingly, it could happen again Sunday, and perhaps even Monday as well.
The record for the number of consecutive days at 125 degrees or higher is 10, set in 1913 (June 28-July 5). This year, Death Valley hit 126 on July 7 and will likely continue that stretch of days with 125-plus temperatures through Tuesday. This would be eight straight days, which would be the second-longest streak in recorded history (tying eight days in 2013). """
```

```python
s = pd.Series(article)
entities = texthero.named_entities(s)[0]
```
