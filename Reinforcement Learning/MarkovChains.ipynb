# Markov Chains

1. **Importing Modules**

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["animation.html"] = "html5"
import matplotlib.animation
```

2. **Defining States of the Markov Chain**: There are three states the weather can take: sunny, cloudy, or rainy. 

```python
states=['sunny','cloudy','rainy']
```

3. **Transition Matrix**: Defines how the state probabilities change from one day to the next:

Screen Shot 2021-05-02 at 4.18.53 PM.png

```python
P=np.array([[0.5,0.5,0],[0.25,0.5,0.25],[0.65,0.2,0.15]])
```

```python
P
```

4. **Define initial state**: Let us start with a sunny day and see how the weather changes everyday

```python
weather=[]
state_array=[]
day1='sunny'
```

```python
weather.append(day1)
state_array.append(0)
```

5. **Simulate Weather**: We do a Monte Carlo simulation to see how weather changes everyday:

```python
for i in range(50):
  state_now=states.index(weather[i])
  state_next=np.random.choice(list(range(len(states))),p=P[state_now,:])
  weather.append(states[state_next])
  state_array.append(state_next)
  print('Day',i+1,': ',weather[i+1])
```

The code below helps us animate the weather simulation:

```python
t=list(range(51))
fig, ax = plt.subplots()
l, = ax.step(t,state_array)
ax.set_yticks([0,1,2])
ax.set_yticklabels(['sunny','cloudy','rainy'])
ax.set_xlabel('Day')
animate = lambda i: l.set_data(t[:i], state_array[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))
```
